## Question 2: Pointers and References (1 point)

Examine the following Python code carefully:

```python
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
new_array = check_array(original_array)

print(original_array)
print(new_array)
```

What does this code output? Explain **why** it produces that output, using the concepts of pointers and references in Python.

When you pass `original_array` to the function, Python doesn't copy the list — it passes a **reference** (basically a pointer) to the same list in memory.

So inside the function, `input` is just another pointer pointing to the **same memory location** as `original_array`. When the function sets `input[idx] = None`, it's following that pointer and modifying the actual list in memory.

By the time the function returns, the original list is already `[None, None]`. Then `new_array` gets assigned the return value — which is that same pointer — so now you have **three names** (`original_array`, `input`, `new_array`) all pointing to the same object.

That's why both print `[None, None]`.

## Question 3: Algorithm Analysis (2 points)

Consider the following algorithm, which finds the contiguous subarray with the largest sum:

```python
max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray
print(max_sum)
```

Analyze the efficiency of this algorithm:

1. Is this algorithm polynomial? Justify your answer.
2. What is the runtime of this algorithm in **Big-O** notation? Show your reasoning.

Yes. A polynomial algorithm is one where the runtime grows as some power of the input size n. This algorithm has two nested loops, both depending on n, so the runtime grows as n² — which is polynomial.

The outer loop runs n times. For each iteration of the outer loop, the inner loop runs n - i times. So the total number of operations is:
n + (n-1) + (n-2) + ... + 1 = n(n+1)/2
When you drop the constants and lower-order terms, that gives you:
O(n²)
So for an array of size 10, you do ~50 operations. For size 1000, you do ~500,000. The work grows quadratically with the size of the input.
